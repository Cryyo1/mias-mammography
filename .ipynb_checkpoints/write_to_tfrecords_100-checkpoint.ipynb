## Verify that the data was written correctly

```python
# make sure everything was written properly by reading it back out
def read_and_decode_single_example(filenames):
    filename_queue = tf.train.string_input_producer(filenames, num_epochs=1)
    
    reader = tf.TFRecordReader()
    
    _, serialized_example = reader.read(filename_queue)
    # The serialized example is converted back to actual values.
    # One needs to describe the format of the objects to be returned
    features = tf.parse_single_example(
        serialized_example,
        features={
            # We know the length of both fields. If not the
            # tf.VarLenFeature could be used
            'label': tf.FixedLenFeature([], tf.string),
            'image': tf.FixedLenFeature([], tf.string)
        })
    
    # now return the converted data
    label = tf.decode_raw(features['label'], tf.uint8)
    image = tf.decode_raw(features['image'], tf.uint8)
    image = tf.reshape(image, [640, 640, 1])
    label = tf.reshape(label, [640, 640, 1])
    
    return label, image
```

```python
label, image = read_and_decode_single_example(["data/training13_0.tfrecords"])

#images_batch, labels_batch = tf.train.shuffle_batch([image, label], batch_size=16, capacity=2000, min_after_dequeue=1000)
images_batch, labels_batch = tf.train.batch([image, label], batch_size=16, capacity=2000)

# labels_batch = tf.cast(labels_batch, dtype=tf.int1)

la_sum = tf.reduce_sum(tf.cast(labels_batch, dtype=tf.int16), axis=[1,2,3])
image_predictions = tf.cast(tf.greater(la_sum, (640*640//850)),dtype=tf.uint8)

# print(la_sum.shape)
# la_sum2 = tf.reduce_mean(la_sum, axis=-1)
# la_max = tf.reduce_max(labels_batch, axis=[2,3])
# print(la_sum2.shape)
# la_mean = tf.reduce_mean(labels_batch, axis=[1,2,3])

global_step = tf.Variable(0, trainable=False)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    sess.run(tf.local_variables_initializer())
    
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(coord=coord)
    for j in range(1):
        la_b, im_b, bar = sess.run([labels_batch, images_batch, image_predictions])
        
        for i in range(len(la_b)):
            f, ax = plt.subplots(1, 2, figsize=(10, 5))
            
            ax[0].imshow(im_b[i].reshape([640,640]))
            ax[0].set_title(str(i) + " (IM) - ")
            
            ax[1].imshow(la_b[i].reshape([640,640]))
            ax[1].set_title(str(i) + " (LA) - " + str(bar[i]))
            plt.show()
            
    coord.request_stop()
    
    # Wait for threads to stop
    coord.join(threads)
```
